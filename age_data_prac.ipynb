```python
import csv
f=open('age_data.csv','r')
data=csv.reader(f)
result=[]
for row in data:
    if '평창동' in row[0]:
        for i in row[3:]:
            i =i.replace(',','')
            result.append(int(i))
print(result)
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rc('font',family='Malgun Gothic')
plt.title('인구 구조')
plt.plot(result)
plt.show()
```
