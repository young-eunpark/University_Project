```python
import csv
import matplotlib.pyplot as plt
f=open('temp_data.csv')
data=csv.reader(f)
next(data)
max_temp=-999
max_date=''
result=[]
for row in data:
    if row[-1] !='': #생일을 임의의 날자로 변경
        if row[0].split('-')[1]=='03' and row[0].split('-')[2]=='21':
            result.append(float(row[-1]))
            if row[-1]=='':
                row[-1]=-999
            row[-1]=float(row[-1])
            if max_temp<row[-1]:
                max_date=row[0]
                max_temp=row[-1]
print("생일 중에 온도가 가장 높았던 날 : "+str(max_date))
print("생일 중에 가장 높았던 온도 : "+str(max_temp))

plt.title("생일 중 최고온도 그래프")
plt.plot(result,'blue')
plt.show()
```
